```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, irfft
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
ekg = np.genfromtxt('ECG Data 1a.csv',delimiter=',').T
```

```python
ekg[0]
```

```python
plt.plot(ekg[1])
```

```python
fft_ekg = rfft(ekg[1])
# fft_ekg[0] = 0.0
fft_ekg[300:] = 0.0

hp = irfft(fft_ekg)
plt.plot(hp)
```

```python
z = list(zip(np.arange(len(hp))/500.0, hp))
```

```python
z.sort(key=lambda x: x[1], reverse=True)
```
