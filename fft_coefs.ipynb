```python
import numpy as np
```

```python
RATE = 44100
WINDOW = 8192
NOTES=[((' ABb B CDb DEb E FGb GAb'[2*i%24:2*i%24+2]+str(i//12)).strip(),27.5*2**(i/12)) for i in range(120)]
FFT_C=[(RATE/WINDOW) * i for i in range(WINDOW//2 + 1)]
```

```python
NOTES
```

```python
notes_to_fft = []
counted = set()
for name,freq in NOTES:
    harmonic = freq
    harmonic_list = []
    while harmonic < 22050 and int(harmonic//freq) <= 16:
        min_q = 2.0
        append = None
        for c in FFT_C:
            if c > harmonic and c/harmonic < min_q:
                min_q = c/harmonic
                append = c
            elif c <= harmonic and c != 0 and harmonic/c < min_q:
                min_q = harmonic/c
                append = c
        harmonic_list.append((int(append/(RATE/WINDOW)), append))
        harmonic += freq
        counted.add(int(append/(RATE/WINDOW)))
    notes_to_fft.append((name, freq, harmonic_list))
    
```

```python
not_counted = set()
for i in range((WINDOW//2) + 1):
    if i not in counted:
        print(i)
        not_counted.add(i)
```
